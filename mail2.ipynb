```python
import win32com.client as win32
import yfinance as yf
from datetime import date, datetime, timedelta
import requests
import os

# 創建 Outlook 應用程式的實例
outlook_app = win32.Dispatch('Outlook.Application')

# 創建郵件項目
mail_item = outlook_app.CreateItem(0)

# 獲取今天的日期
today = date.today().strftime("%Y/%m/%d")

# 設定郵件的主旨
mail_item.Subject = f"每日匯率報告==> - {today}"

# 使用 Yahoo Finance 套件獲取每日匯率
usdtwd = yf.Ticker("USDTWD=X")
usdcny = yf.Ticker("USDCNY=X")
usdjpy = yf.Ticker("USDJPY=X")
twdjpy = yf.Ticker("TWDJPY=X")
exchange_rate1 = 1 / (twdjpy.history(period="1d").iloc[0]['Close'])
exchange_rate2 = usdjpy.history(period="1d").iloc[0]['Close']
exchange_rate3 = usdtwd.history(period="1d").iloc[0]['Close']
exchange_rate4 = usdcny.history(period="1d").iloc[0]['Close']

# 使用 HTML 組織郵件內容
html_body = f"""
<html>
<head>
     <style>
        table {{
            border-collapse: collapse; 
            width: 500px;
        }}
        body{{
            font-family: arial,"Microsoft JhengHei","微軟正黑體",sans-serif !important;
        }}
        th, td {{
            border: 1px solid lightgrey;
            padding: 8px;
        }}
        th {{
            background-color:#00558c;
            color: white;
        }}         
    </style>
</head>
<body>
    <table   >
        <tr >
            <th>{today}</th>
            <th>現行匯率<br>(寫給客戶看的)</th>
        </tr>
        <tr style="font-weight: bold;">
            <td align="center" valign="center">日幣：台幣</td>
            <td align="right">{exchange_rate1}</td>
        </tr>
        <tr style="background-color: #fffafa;font-weight: bold;">
            <td align="center" valign="center">美金：日幣</td>
            <td align="right">{exchange_rate2}</td>
        </tr>
        <tr style="font-weight: bold;">
            <td align="center" valign="center">美金：台幣</td>
            <td align="right">{exchange_rate3}</td>
        </tr>
        <tr style="background-color: #fffafa;font-weight: bold;">
            <td align="center" valign="center">美金：人民幣</td>
            <td align="right">{exchange_rate4}</td>
        </tr>
    </table>
</body>
</html>
"""

# 設定郵件主體為 HTML 內容
mail_item.HTMLBody = html_body

# 設定收件人
mail_item.To = "chenxuan8891@gmail.com"

# 下載附件
attachment_url = "https://www.cbc.gov.tw/Public/Data/economic/statistic/fx/日資料 (2002年迄今).xls"
last_month = datetime.now().replace(day=1) - timedelta(days=1)
attachment_path = f"C:\\Users\\chenxuan\\Downloads\\Average Rate {last_month.strftime('%Y.%m')}.xls"  # 設定本地儲存路徑
attachment_display_name = f"Average Rate {last_month.strftime('%Y.%m')}.xls"
attachment_type = 1  # 1 表示檔案附件

response = requests.get(attachment_url)
with open(attachment_path, "wb") as file:
    file.write(response.content)

# 添加附件到郵件
mail_item.Attachments.Add(attachment_path, attachment_type, 1, attachment_display_name)

# 發送郵件
mail_item.Send()

# 刪除本地的 XLS 檔案
os.remove(attachment_path)

print("郵件已發送")
```
